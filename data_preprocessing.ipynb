```python
import pandas as pd
```

```python
df1 = pd.read_csv('amazon.csv', index_col=0)
df2 = pd.read_csv('flipkart.csv', index_col=0)
```

```python
df1['disc'] = df1['disc'].str.replace('% off)', "")
df1['disc']= df1['disc'].str.replace("(", "")
df1['stars'] = df1['stars'].str.replace("out of 5 stars", "")
df1['sales'] = df1['sales'].str.replace(",", "")
df1['price'] = df1['price'].str.replace(",", "")
df2['disc'] = df2['disc'].str.replace('% off)', "")
df2['disc']= df2['disc'].str.replace("(", "")
df2['stars'] = df2['stars'].str.replace("out of 5 stars", "")
df2['sales'] = df2['sales'].str.replace(",", "")
df2['price'] = df2['price'].str.replace(",", "")
```

```python
df1['sales'] = pd.to_numeric(df1['sales'])
df1['stars'] = pd.to_numeric(df1['stars'])
df1['price'] = pd.to_numeric(df1['price'])
df1['disc'] = pd.to_numeric(df1['disc'])
df2['sales'] = pd.to_numeric(df2['sales'])
df2['stars'] = pd.to_numeric(df2['stars'])
df2['price'] = pd.to_numeric(df2['price'])
df2['disc'] = pd.to_numeric(df2['disc'])
```

```python
df1.info()
df2.info()
```

```python
df1dd = df1.drop_duplicates(subset = ['title', 'price', 'stars', 'sales', 'disc'], keep=False)
```

```python
df2dd = df2.drop_duplicates(subset = ['title', 'price', 'stars', 'sales', 'disc'], keep=False)
```
