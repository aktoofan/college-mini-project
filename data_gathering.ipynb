```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def amazon_search_products(query, num_pages=10):
    results = []

    for page in range(1, num_pages + 1):
        url = f"https://www.amazon.in/s?k={query}&page={page}"
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0;Win64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
        }
        response = requests.get(url, headers=headers)
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')

        product_containers = soup.find_all("div", class_="sg-col-inner")
        for container in product_containers:
            title_element = container.find("span", class_="a-size-medium a-color-base a-text-normal")
            price_element = container.find("span", class_="a-price-whole")
            star_element = container.find("span", class_="a-icon-alt")
            sales_element = container.find("span", class_="a-size-base s-underline-text")
            disc_element = container.select_one("span:nth-child(3)")

            if title_element is None or price_element is None or star_element is None or sales_element is None or disc_element is None :
                continue

            title = title_element.text.strip()
            price = price_element.text.strip()
            stars = star_element.text.strip()
            sales = sales_element.text.strip()
            disc = disc_element.text.strip()

            result = {
                'title': title,
                'price': price,
                'stars': stars,
                'sales': sales,
                'disc': disc
            }
            results.append(result)

    return results

q = input("Enter the product name: ")
product_search_results = amazon_search_products(q)

if product_search_results:
    df1 = pd.DataFrame(product_search_results)
    print(df1)
else:
    print("No results found.")
```

```python
def flipkart_search_products(query, num_pages=10):
    results = []

    for page in range(1, num_pages + 1):
        url = f"https://www.flipkart.com/search?q={query}&page={page}"
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0;Win64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
        }
        response = requests.get(url, headers=headers)
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')

        product_containers = soup.find_all("div", class_="_4ddWXP")
        for container in product_containers:
            title_element = container.find("a", class_="s1Q9rs")
            price_element = container.find("div", class_="_30jeq3")
            star_element = container.find("div", class_="_3LWZlK")
            sales_element = container.find("span", class_="_2_R_DZ")
            disc_element = container.select_one("div:nth-child(1)")

            if title_element is None or price_element is None or star_element is None or sales_element is None or disc_element is None:
                continue

            title = title_element.text.strip()
            price = price_element.text.strip()
            stars = star_element.text.strip()
            sales = sales_element.text.strip()
            disc = disc_element.text.strip()
            
            result = {
                'title': title,
                'price': price,
                'stars':stars,
                'sales':sales,
                'disc':disc
            }
            results.append(result)

    return results

q = input("Enter the product name: ")
product_search_results = amazon_search_products(q)

if product_search_results:
    df2 = pd.DataFrame(product_search_results)
    print(df2)
else:
    print("No results found.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

results = []

m1 = [0, 1, 2, 3, 4, 5, 6]
for page in m1:  # Scraping the first 5 pages
    url = f"https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start={page}1&ref_=adv_prv"
    response = requests.get(url)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    movie_containers = soup.find_all("div", class_="lister-item-content")

    for i in movie_containers:
        movie_name_element = i.select_one("h3 > a")
        year_element = i.find("span", class_="lister-item-year text-muted unbold")
        time_element = i.find("span", class_="runtime")
        genre_element = i.find("span", class_="genre")
        rating_element = i.select_one("strong")
        metascore_element = i.find("span", class_="metascore favorable")

        if (
            movie_name_element is None
            or year_element is None
            or time_element is None
            or genre_element is None
            or rating_element is None
        ):
            continue

        movie = movie_name_element.text.strip()
        year = year_element.text.strip()
        time = time_element.text.strip()
        genre = genre_element.text.strip()
        rating = rating_element.text.strip()
        metascore = metascore_element

        result = {
            'movie': movie,
            'year': year,
            'time': time,
            'genre': genre,
            'rating': rating,
            'metascore': metascore
        }

        results.append(result)

df2 = pd.DataFrame(results)
df2
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

results = []

for i in range(1):
    url = "https://www.flipkart.com/search?q=iphones&page="+str(i)
    response = requests.get(url)
    html_content = response.text
    soup = BeautifulSoup(html_content,"html.parser")

    #product names

    box1 = soup.find_all("div",class_="_2kHMtA")
    for box in box1:
        name = box.find("div",class_="_4rR01T")
        price = box.find("div",class_="_30jeq3 _1_WHN1")
        des = box.find("ul",class_="_1xgFaf")
        rat = box.find("div",class_="_3LWZlK")
        off = box.find("div",class_="_3Ay6Sb")
        
        names = name.text.strip()
        prices = price.text.strip()
        Desc = des.text.strip()
        ratings = rat.text.strip()
        offers = off.text.strip()
        
        result = {"Product Name":names,"Product Price":prices,"Description":Desc,"Rating":ratings,"Offers":offers}
        results.append(result)

data = pd.DataFrame(results)
print(data)
```

```python
data.to_csv("iphone_flipkart.csv")
```
